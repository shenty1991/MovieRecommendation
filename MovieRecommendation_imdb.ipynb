```python
df_principals = pd.read_table('./imdb_data/principals.tsv')
df_principals.info()
df_principals.head()
```

```python
df_principals = df_principals[df_principals['category'].isin(['director','actor','actress'])]
df_principals.info()
df_principals.head()
```

```python
df_principals['tconst']=df_principals['tconst'].apply(lambda s: int(s[2:]))
df_principals.head()
```

```python
df_principals = df_principals[df_principals['tconst'].isin(df_links_popular['imdbId'])]
df_principals.info()
df_principals.head()
```

```python
df_merged = pd.merge(df_popular,df_links_popular,on='movieId')
```

```python
df_merged = df_merged.drop(columns=['tmdbId','Unnamed: 0'])
df_merged.info()
df_merged.head()
```

```python
df_principals[df_principals['tconst']==114709]
```

```python
def findDirector(imdbId):
    df_imdbId = df_principals[df_principals['tconst']==imdbId]
    directorId = df_imdbId[df_imdbId['category']=='director']
    return list(directorId['nconst'].values)

findDirector(114709)
```

```python
def findActor(imdbId):
    df_imdbId = df_principals[df_principals['tconst']==imdbId]
    ActorIds = df_imdbId[df_imdbId['category']=='actor']
    return list(ActorIds['nconst'].values)

findActor(114709)
```

```python
def findActress(imdbId):
    df_imdbId = df_principals[df_principals['tconst']==imdbId]
    ActressIds = df_imdbId[df_imdbId['category']=='actress']
    return list(ActressIds['nconst'].values)

findActress(114709)
```

```python
df_merged['director']=df_merged['imdbId'].apply(findDirector)
df_merged['actor']=df_merged['imdbId'].apply(findActor)
df_merged['actress']=df_merged['imdbId'].apply(findActress)
df_merged.head()
```

```python
df_names=pd.read_table('./imdb_data/names.tsv')
df_names.info()
df_names.head()
```

```python
crewIdlist=[]
for id in df_merged['imdbId']:
    directors = findDirector(id)
    actors = findActor(id)
    actrice = findActress(id)
    for i in range(len(directors)):
        crewIdlist.append(directors[i])
    for i in range(len(actors)):
        crewIdlist.append(actors[i])
    for i in range(len(actrice)):
        crewIdlist.append(actrice[i])
    
crewIdlist=list(set(crewIdlist))
        
print(crewIdlist)        
```

```python
df_names=df_names[df_names['nconst'].isin(crewIdlist)]
df_names.info()
df_names.head()
df_names.to_csv('./data/names.csv')
```
